```python
import asyncio
from fastmcp import Client
import nest_asyncio

# Allow nested event loops (needed for notebooks)
nest_asyncio.apply()

client = Client("http://localhost:9010/mcp")
```

```python


async def call_get_state_list():
    async with client:
        result = await client.call_tool("get_state_list_from_enam")
        return result

# Run it safely in Jupyter
result_ = await call_get_state_list()
```

```python
async def call_get_state_list(client):
    async with client:
        result = await client.call_tool("get_commodities_available_for_KA")
        return result

r=await call_get_state_list(client)
```
